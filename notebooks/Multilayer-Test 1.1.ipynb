```python
g = []
for i in range(len(adj)):
    g.append(ig.Graph.Incidence(adj[i].values.tolist()))
```

```python
#name nodes
#keep a counter for j and k, j for red nodes and k for blue
#red nodes use column list, blue nodes use index

#assumes that the names are in order...

for i in range(len(g)):
    

    j = 0
    k = 0
    for l in range(g[i].vcount()):
        if (g[i].vs[l]['type']):
            g[i].vs[l]['label'] = list(adj[i])[j]
            g[i].vs[l]['color'] = 'red'
            j+= 1
        else:
            g[i].vs[l]['color'] = 'blue'
            g[i].vs[l]['label'] = list(adj[i].index.values)[k]
            k+=1 

        
```

```python
df = []
for i in range(len(g)):
    
    df.append(pd.DataFrame({'name': g[i].vs['label'], 
                            'type': g[i].vs['type'], 
                            'betweenness': g[i].betweenness(),
                           'degree': g[i].degree(),
                            'pagerank': g[i].pagerank()
                           }))
```
